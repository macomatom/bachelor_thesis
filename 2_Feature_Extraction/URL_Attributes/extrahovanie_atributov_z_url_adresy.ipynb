```python
import pandas as pd
```

```python
df = pd.read_csv("dataset_final_concat.csv")
df.head()
```

```python
# URL length
df["length_url"] = df['link'].apply(lambda x: len(x))
df.head(50)
```

```python
# hostname length
from urllib.parse import urlparse
df["length_hostname"] = df['link'].apply(lambda x: len(f"{urlparse(x).scheme}://{urlparse(x).netloc}/"))
df.head(50)
```

```python
# number of dots in URL
df["nb_dots"] = df['link'].apply(lambda x: x.count('.'))
df.head(50)
```

```python
# number of hyphens
df["nb_hyphens"] = df['link'].apply(lambda x: x.count('-'))
df.head(50)
```

```python
# number of "@"
df["nb_at"] = df['link'].apply(lambda x: x.count('@'))
df.head(50)
```

```python
# number of question marks
df["nb_qm"] = df['link'].apply(lambda x: x.count('?'))
df.head(50)
```

```python
# number of "&""
df["nb_and"] = df['link'].apply(lambda x: x.count('&'))
df.head(50)
```

```python
# number of "|"
df["nb_or"] = df['link'].apply(lambda x: x.count('|'))
df.head(50)
```

```python
# number of equal signs
df["nb_eq"] = df['link'].apply(lambda x: x.count('='))
df.head(50)
```

```python
# number of underscores
df["nb_underscore"] = df['link'].apply(lambda x: x.count('_'))
df.head(50)
```

```python
# number of tildes "~"
df["nb_tilde"] = df['link'].apply(lambda x: x.count('~'))
df.head(50)
```

```python
# number of percent signs "%"
df["nb_percent"] = df['link'].apply(lambda x: x.count('%'))
df.head(50)
```

```python
# number of slashes "/"
df["nb_slash"] = df['link'].apply(lambda x: x.count('/'))
df.head(50)
```

```python
# number of stars "*"
df["nb_star"] = df['link'].apply(lambda x: x.count('*'))
df.head(50)
```

```python
# number of colons ":"
df["nb_colon"] = df['link'].apply(lambda x: x.count(':'))
df.head(50)
```

```python
# number of commas ","
df["nb_comma"] = df['link'].apply(lambda x: x.count(','))
df.head(50)
```

```python
# number of semicolons ";"
df["nb_semicolon"] = df['link'].apply(lambda x: x.count(';'))
df.head(50)
```

```python
# number of "$"
df["nb_dollar"] = df['link'].apply(lambda x: x.count('$'))
df.head(50)
```

```python
# number of spaces
df["nb_space"] = df['link'].apply(lambda x: x.count(' '))
df.head(50)
```

```python
# number of "www"
df["nb_www"] = df['link'].apply(lambda x: x.count("www"))
df.head(50)
```

```python
# number of "com"
df["nb_com"] = df['link'].apply(lambda x: x.count("com"))
df.head(50)
```

```python
# number of backward slashes "\"
df["nb_dslash"] = df['link'].apply(lambda x: x.count('\\'))
df.head(50)
```

```python
# number of "http" in path
df["http_in_path"] = df['link'].apply(lambda x: x.count("http"))
df.head(50)
```

```python
# URL contains https token
df["https_token"] = df['link'].apply(lambda x: 1 if "https" in x else 0)
df.head(50)
```

```python
# ratio of number of digits to length of url
df["ratio_digits_url"] = df['link'].apply(lambda x: sum(c.isdigit() for c in x) / len(x))
df.head(50)
```

```python
# ratio of number of digits to length of host
df["ratio_digits_host"] = df['link'].apply(lambda x: sum(c.isdigit() for c in x) / len(f"{urlparse(x).scheme}://{urlparse(x).netloc}/"))
df.head(50)
```
