```python
import pandas as pd
```

```python
df = pd.read_csv("rozsireny_dataset.csv")
print(df.shape)
print(df['h1'].count())
df.head()
```

```python
# Nahradanie NaN-ov prazdnymi stringami
df = df.fillna("")
df.head()
```

```python
# Concatenating columns to single one
column_names = ["title", "h1", "h2", "h3", "h4", "h5", "h6", "p", "a", "span"]
df['value'] = df[column_names].agg(' '.join, axis=1)
df.head()
```

```python
# Dropping redundant columns
df.drop(df.columns[6:-1], axis=1, inplace=True)
# df.columns[4:-1]
df.head()
```

```python
df.iloc[0]['value']
```

```python
# Reordering columns
df = df[['index', 'name', 'link', 'lang_eng', 'valid_dir', 'value', 'type']]
df.head()
```

```python
df.to_csv("dataset_concat.csv", index=False)
```
