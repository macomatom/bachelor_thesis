```python
import pandas as pd
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv("../zoznam_url_final.csv")
# df = df.loc[df['type'].isin(['Education', 'Company'])]
df.head(50)
```

```python
# datatype info
df.info()
```

```python
import progressbar
path = "../data_zdrojove_kody_stranok/"

column_names = ["title", "h1", "h2", "h3", "h4", "h5", "h6", "p", "a"]
df = df.reindex(columns=df.columns.tolist() + column_names)

indexes = df["index"]
with progressbar.ProgressBar(max_value=len(indexes)) as bar:
    for i, index in enumerate(indexes):
        try:
            soup = BeautifulSoup(open(f"{path}{index}.txt"))
            for name in column_names:
                my_arr = soup.find(name)
                my_arr = my_arr.find_all(text=True, recursive=False) if my_arr is not None else []
                # print(my_arr)
                if len(my_arr) > 0:
                    content = " ".join(my_arr)
                    df.loc[df["index"] == index, name] = content
            # print("----------------------")
        except PermissionError:
            print(f"Unable to open file \"{path}{index}.txt\"")
        except FileNotFoundError:
            print(f"Unable to open file \"{path}{index}.txt\"")
        bar.update(i)
```
